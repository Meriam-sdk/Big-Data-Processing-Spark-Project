## Part A plotting 1:



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('/content/output1.csv')
df1.columns = ['Date', 'Total Transactions per month']

plt.figure(figsize=(20, 8))  # Set the figure size
plt.xticks(rotation=90, fontsize=12)  # Set the x-axis label rotation and font size
plt.yticks(fontsize=12)  # Set the y-axis label font size
plt.xlabel('Month/Year', fontsize=14)  # Set the x-axis label and font size
plt.ylabel('Total Transactions per month', fontsize=14)  # Set the y-axis label and font size
plt.title('Total Transactions Each Month', fontsize=16)  # Set the plot title and font size

x = df1['Date']
y = df1['Total Transactions per month']
plt.bar(x, y, color='green')

# Display the plot
plt.show()
```

## plotting for gas price 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df3 = pd.read_csv('/content/average_gas_price.csv')
df3.columns = ['Date', 'Average gas price each month']

# Create a new figure and axis object
fig, ax = plt.subplots(figsize=(20, 6))  # Set the figure size

# Plot the data on the axis object
x = df3['Date']
y = df3['Average gas price each month']
ax.plot(x, y, color='green')

# Add gridlines to the plot
ax.grid(True, which='both', linestyle='--', color='grey', alpha=0.5)

# Set the x and y axis labels and font size
ax.set_xlabel('Month/Year', fontsize=14)
ax.set_ylabel('Average gas price each month', fontsize=14)

# Rotate the x-axis tick labels and set font size
plt.xticks(rotation=50, fontsize=12)

# Set the plot title and font size
plt.title('Average Gas Price Each Month', fontsize=16)

# Display the plot
plt.show()
```
